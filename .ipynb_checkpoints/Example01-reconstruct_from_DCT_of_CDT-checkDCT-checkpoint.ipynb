```python
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import rcParams
from scipy.signal import savgol_filter
from scipy.ndimage.filters import gaussian_filter
# rcParams.update({'font.size': 12})
# import matplotlib as mpl
# label_size = 20
# mpl.rcParams['xtick.labelsize'] = label_size 
```

```python
from skimage.io import imread
# from skimage.data import shepp_logan_phantom
from skimage.transform import radon, rescale

image = imread('SheppLogan_Phantom.png')
image = rescale(image, scale=0.5, mode='reflect', multichannel=False)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Image")
ax1.imshow(image, cmap=plt.cm.Greys_r)

theta = np.linspace(0., 180., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta, circle=True)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')

fig.tight_layout()
plt.show()
```

```python
I1=sinogram[:, 0]
I1 = gaussian_filter(I1, sigma=5.0)
# I1 = I1[40:-40]
# I1 = np.pad(I1, [50, 50], mode='constant')
```

```python
N=len(I1)
x=np.arange(N)
x0 = np.linspace(0, 1, N)
#I0= np.ones(N)
sig = 0.1
I0 = np.exp(-np.power(x0 - 1/2, 2.) / (2 * np.power(sig, 2.)))
```

```python
epsilon = 1e-3
I0 = abs(I0) + epsilon
I0 = I0/I0.sum()

plt.plot(x0,I0)
plt.show()
I1 = I1 + 50
I1 = abs(I1) + epsilon
I1 = I1/I1.sum()
```

```python
from optrans.continuous.cdt import CDT

# Create a CDT object
cdt = CDT()

# Compute the forward transform
I1_hat, f1_hat, xtilde = cdt.forward(x0, I0, x, I1, rm_edge=False)
# f1_hat = f1_hat[25:-25]

I1_recon, I1_recon_f1 = cdt.inverse(f1_hat, I0, x)

# Plot I1 and I1_hat
# fontSize=14
fig, ax = plt.subplots(1, 3, sharex=False, sharey=False, figsize=(15,5))
ax[0].plot(x,I1, 'b-',linewidth=2)
ax[0].set_title('Original signal')
#ax[0].set_xticks([])
#ax[0].set_yticks([])

ax[1].plot(x0,f1_hat, 'b-',linewidth=2)
ax[1].set_title('CDT')
#ax[1].set_xticks([])
#ax[1].set_yticks([])

ax[2].plot(x,I1_recon, 'b-',linewidth=2)
ax[2].set_title('Signal_recon')
#ax[2].set_xticks([])
#ax[2].set_yticks([])

plt.show()
```

# Calculate DCT coefficients

Using DCT of image:

```python
from scipy.fftpack import dct, idct, dst, idst
f1_hat_recon = None

def dct_recons(x, num_coefficents):
    N = len(x)
    t = np.linspace(0, 1, N)
    y = dst(x, norm='ortho')
    th = np.sort(np.abs(y))[::-1][num_coefficents-1]
    y[np.abs(y)<th] = 0
    yr = idst(y, norm='ortho')
    return yr


def cdt_recons(x, f1_hat, N, I0):
    I1_hat = (f1_hat - x)*np.sqrt(I0)

    I1_hat_recon = dct_recons(I1_hat, N)
   
    f1_hat_recon = x + I1_hat_recon/np.sqrt(I0)
    
    #f1_hat_recon = (f1_hat_recon - f1_hat_recon.min())/(f1_hat_recon.max()- f1_hat_recon.min())
    #f1_hat_recon = f1_hat_recon*f1_hat.max()

#     ## CUT
#     N = len(f1_hat_recon)
#     size = 25
#     f1_hat_recon = f1_hat_recon[size:-size]
#     f1_hat_recon = np.interp(np.linspace(0, 1, N), np.linspace(0, 1, N-50), f1_hat_recon)
    
#     f1_hat_recon = np.pad(f1_hat_recon, [size, size], mode='wrap')
#     plt.plot(f1_hat_recon)

    I1_recon, I1_recon_f1 = cdt.inverse(f1_hat_recon, I0, x)


    return I1_hat, I1_hat_recon, f1_hat_recon, I1_recon_f1


nth_max = len(I1)
#nth_max = 4  # select number of largest DCT coefficients to use
I_recon = dct_recons(I1, nth_max)
I1_hat, I1_hat_recon, f1_hat_recon, I1_recon = cdt_recons(x, f1_hat, nth_max, I0)
```

Using DCT of CDT:

```python
MSE_dct = []
MSE_cdt = []

for N in range(1, 113):
    I_recon = dct_recons(I1, N)
    I1_hat, I1_recon, f1_hat_recon, I1_recon = cdt_recons(x0, f1_hat, N, I0)
    MSE_dct.append(np.mean((I1-I_recon)**2))
    MSE_cdt.append(np.mean((I1-I1_recon)**2))
print(np.argmax(MSE_cdt))
print(np.max(MSE_cdt))
```

```python
plt.Figure(figsize=(4, 3), dpi=200)
plt.plot(MSE_dct, label='DCT reconstruction', color='blue')
plt.plot(MSE_cdt, label='CDT-DCT reconstruction', color='green')
plt.ylim([0, 0.0001])
plt.ylabel('Mean Square Error')
plt.xlabel('Number of Coefficients')
plt.legend()
plt.savefig('mse.png', dpi=200)
plt.show()
print(np.argmax(MSE_cdt[10:]))
```
